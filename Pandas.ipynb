```python
#program to find greatest number among two number using function.
def great():
    a,b=10,20
    if a>b:
        print("a is greater than b")
    elif a==b:
        print("a and b have same value")
    else:
        print("b is greater than a")

great()
```

```python
#program to find greatest number among three numbers using function.
def three():
    a,b,c=10,20,30
    if b>a and b>c:
        print("b is greater than a and c.")
    elif c>a and c>b:
        print("c is greater than a and b.")
    else:
        print("a is greater than b and c.")
three()
```
